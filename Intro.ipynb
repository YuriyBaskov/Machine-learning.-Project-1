<h2>1. 5 примеров применения методов МО в жизни:</h2>

1. Рекомендательные системы (например, Netflix, Amazon)
Преимущество: Персонализированный пользовательский опыт, увеличение вовлеченности и продаж.

2. Медицинская диагностика (анализ рентгеновских снимков)
Преимущество: Более точная и быстрая диагностика, снижение нагрузки на врачей.

3. Прогнозирование пробок (Google Maps, Яндекс.Карты)
Преимущество: Оптимизация маршрутов, экономия времени пользователей.

4. Фильтрация спама в электронной почте
Преимущество: Автоматическая защита пользователей от нежелательных сообщений.

5. Автономные автомобили
Преимущество: Повышение безопасности дорожного движения и удобства передвижения.

<h2>2. Классификация задач из таблицы:</h2>

1. Прогноз цены на дом - контролируемое обучение (регрессия)

2. Кредитный скоринг - контролируемое обучение (классификация)

3. Прогноз приема лекарств - контролируемое обучение (регрессия/классификация)

4. Выбор лекарства - контролируемое обучение (классификация)

5. Сегментация клиентов - неконтролируемое обучение (кластеризация)

<h2>3. Разница между мультиклассом и мультиметкой:</h2>

Мультикласс: Каждый объект принадлежит ровно одному классу из множества (например, фрукт может быть только яблоком или грушей).

Мультиметка: Каждый объект может иметь несколько меток одновременно (например, статья может быть помечена как "политика" и "экономика" одновременно).

<h2>4. Пример с ценами на жилье:</h2>

Это регрессионная проблема, так как целевая переменная (цена) - непрерывная величина. Регрессионную проблему можно свести к классификации, если дискретизировать целевую переменную (например, разделить цены на диапазоны: "дешевый", "средний", "дорогой"), но это приведет к потере информации.
